```python
import pandas as pd
import pytrends
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from pytrends.request import TrendReq
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list =["ESG"]
pytrends.build_payload(kw_list, geo ='GR')
```

```python
df=pytrends.interest_over_time()
```

```python
df.to_csv("greekGoogleTrendEsg.csv",sep=';')
df.head()
df['ESG'].dtype
```

```python
ax=df['ESG'].plot(figsize=(18,6),title='Google Trend for ESG in Greece')
ax.set(label='Interest over time')
```
